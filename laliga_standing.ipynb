```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import pandas as pd
```

```python
URL = "https://www.soccerstats.com/latest.asp?league=spain"
response = requests.get(URL)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
table = soup.select("#btable")[9]
```

```python
links = table.find_all("a")
```

```python
links = [l.get("href") for l in links]
```

```python
links = [l for l in links if "team.asp" in l]
```

```python
links
```

```python
team_urls = [f"https://www.soccerstats.com/{l}" for l in links]
team_urls
```

```python
team_url = team_urls[0]
data = requests.get(team_url)
```

```python
matches = pd.read_html(data.content, match="2.5+")
matches = matches[0].dropna()
```

```python
matches[['Final Score', 'Goal Details']] = matches[2].str.extract(r'(\d+\s*-\s*\d+)(.*)')


matches
```

```python
matches
```

```python
matches.dropna(inplace=True)
```

```python
final_score = matches.pop("Final Score")
matches.insert(2, "Final Score", final_score)
```
